```python
import pandas as pd
```

```python
df=pd.read_excel('Data Model - Pizza Sales.xlsx')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# Change the data type and split
df['order_time']=df['order_time'].astype('string')
df[['hour','minute', 'second']]= df['order_time'].str.split(":",expand=True)
```

```python
df['hour']=df['hour'].astype('object')
df.info()
```

```python
# To check duplicates
df.groupby('order_id')['order_id'].agg("count")
```

```python
# To check duplicates and typos
df.groupby('pizza_name')['pizza_name'].agg("count")
```

```python
# Drop columns which not necessary and make sense 
df1=df.drop(['order_id','minute', 'second'], axis=1)
df1
```

```python
df1.info()
```

```python
# Export transformed table to .xlxs 
df1.to_excel(f"Data Model - Pizza Sales.xlsx(Clean Set).xlsx",index=False)
```
